# Importing_Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data_Collection

```python
uber_data= pd.read_csv("https://raw.githubusercontent.com/Karan-Malik/UberDataAnalysis/master/uber-raw-data-apr14.csv")
uber_data.head()
```

# Activating_Date/Time

```python
uber_data["Date/Time"]= uber_data["Date/Time"].map(pd.to_datetime)
uber_data
```

# Extracting days,hours & weekdays columns;

```python
uber_data["days"]= uber_data["Date/Time"].apply(lambda x:x.day)
uber_data["hour"]= uber_data["Date/Time"].apply(lambda x:x.hour)
uber_data["week_days"]= uber_data["Date/Time"].apply(lambda x:x.weekday())
```

```python
uber_data
```

# Analysing Uber_trip by the days;

```python
sns.set(rc={"figure.figsize":(15,13)})
sns.distplot(uber_data["days"])
```

# Analysing Uber_trip by the hours;

```python
sns.set(rc={"figure.figsize":(15,13)})
sns.distplot(uber_data["hour"])
```

# Analysing Uber_trip by the week_days;

```python
sns.set(rc={"figure.figsize":(15,13)})
sns.distplot(uber_data["week_days"])
```

# Analysing trip using correlation_plot

```python
df= uber_data.groupby(["week_days","hour"]).apply(lambda x:len(x))
df= df.unstack()
sns.heatmap(df,annot=True)
```

# Analysing Density of trip

```python
uber_data.plot(kind="scatter",x="Lat",y="Lon",s=uber_data["days"],label="Uber_trip",figsize=(12,10))
```
